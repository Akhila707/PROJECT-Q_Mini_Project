```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

def generate_quantum_bit():
    # 1. Create a circuit with 1 qubit and 1 classical bit
    circuit = QuantumCircuit(1, 1)
    
    # 2. Put the qubit into a 50/50 superposition using the Hadamard gate
    circuit.h(0)
    
    # 3. Measure the qubit
    circuit.measure(0, 0)
    
    # 4. Run the circuit for exactly 1 shot
    simulator = AerSimulator()
    result = simulator.run(circuit, shots=1, memory=True).result()
    
    # 5. Extract the bit string result ('0' or '1')
    measured_bit = result.get_memory()[0]
    return int(measured_bit)

# Generate a sequence of 20 quantum random bits
quantum_bits = [generate_quantum_bit() for _ in range(20)]

print("Quantum Random Numbers")
print(quantum_bits)
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt

def generate_quantum_bit():
    circuit = QuantumCircuit(1, 1)
    circuit.h(0)
    circuit.measure(0, 0)
    simulator = AerSimulator()
    result = simulator.run(circuit, shots=1, memory=True).result()
    return int(result.get_memory()[0])

# Generate 100 quantum random bits for statistics
quantum_dataset = [generate_quantum_bit() for _ in range(100)]
zeros = quantum_dataset.count(0)
ones = quantum_dataset.count(1)

# Plotting the results
plt.figure(figsize=(6, 4))
plt.bar(['0', '1'], [zeros, ones], color=['#06B6D4', '#8B5CF6'], width=0.5)
plt.xlabel('Quantum States')
plt.ylabel('Frequencies')
plt.title('Quantum Randomness Distribution')

# Display exact count values over the bars
plt.text(0, zeros + 1, str(zeros), ha='center', fontweight='bold')
plt.text(1, ones + 1, str(ones), ha='center', fontweight='bold')

plt.show()
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import numpy as np

def generate_quantum_bit():
    circuit = QuantumCircuit(1, 1)
    circuit.h(0)
    circuit.measure(0, 0)
    simulator = AerSimulator()
    result = simulator.run(circuit, shots=1, memory=True).result()
    return int(result.get_memory()[0])

quantum_dataset = [generate_quantum_bit() for _ in range(100)]
zeros = quantum_dataset.count(0)
ones = quantum_dataset.count(1)

print("Statistical Evaluation")
print(f"Zeros: {zeros}")
print(f"Ones: {ones}")
print("Mean:", np.mean(quantum_dataset))
print("Standard Deviation:", np.std(quantum_dataset))
```
